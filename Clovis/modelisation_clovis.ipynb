```python
X = X.drop(columns=["nom_departement"])
X.head()
```

## Decision Tree Classifier

```python
tree = DecisionTreeClassifier()
```

```python
tree
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
X_train
```

```python
region_train = pd.get_dummies(X_train["nom_region"])
```

```python
X_train.drop(columns=["nom_region"], inplace=True)
```

```python
X_train_enc = pd.concat([X_train, region_train], axis=1)
```

```python
region_test = pd.get_dummies(X_test["nom_region"])
X_test.drop(columns=["nom_region"], inplace=True)
X_test_enc = pd.concat([X_test, region_test], axis=1)
```

```python
parameters = {'min_samples_split': [100,500,1000,5000,10000], 
               'min_samples_leaf':[50,250,500,2500,5000],
               'random_state':[42]
               }
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf = GridSearchCV(tree, parameters)
```

```python
X_train_enc.head()
```

```python
clf.fit(X_train_enc, y_train) 
```

```python
importance = clf.best_estimator_.feature_importances_
importance
```

```python
for feat, importance in zip(X_train_enc.columns, clf.best_estimator_.feature_importances_):
    print('feature: {f}, importance: {i}'.format(f=feat, i=importance))
```

```python
print(clf.best_params_) 
grid_predictions = clf.predict(X_test_enc) 
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, grid_predictions)) 
```

# Prédiction Autres

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
df_model = df.loc[(df["Type local"] != "Autres")]
```

```python
tree = DecisionTreeClassifier(min_samples_leaf= 50, min_samples_split=100, random_state= 42)
```

```python
X = df_model.drop(columns=["Valeur fonciere", "Type local", "year"])
X = X.drop(columns=["nom_departement"])
y = df_model.iloc[:,2]
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33, random_state=42)
```

```python
region_train = pd.get_dummies(X_train["nom_region"])
X_train.drop(columns=["nom_region"], inplace=True)
X_train_enc = pd.concat([X_train, region_train], axis=1)
region_test = pd.get_dummies(X_test["nom_region"])
X_test.drop(columns=["nom_region"], inplace=True)
X_test_enc = pd.concat([X_test, region_test], axis=1)
```

```python
X_train_enc.head()
```

```python
tree.fit(X_train_enc, y_train)
```

```python
y_pred = tree.predict(X_test_enc) 
```

```python
print(classification_report(y_test, y_pred)) 
```
