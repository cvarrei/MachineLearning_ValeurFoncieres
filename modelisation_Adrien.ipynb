```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("Data/df6.csv", delimiter="\t", nrows=500000)
```

```python
df.drop(columns=["Unnamed: 0"],inplace=True)
```

```python
from sklearn.preprocessing  import LabelEncoder

# Créez une instance de LabelEncoder
label_encoder = LabelEncoder()

# Appliquez le LabelEncoder à la colonne 'Catégorie'
df['Type local encoded'] = label_encoder.fit_transform(df['Type local'])
```

```python
df_train=df[df["Type local"]!="Autres"]
df_test=df[df["Type local"]=="Autres"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df[["Valeur fonciere","Code postal","Nombre de lots","Surface reelle bati","Nombre pieces principales",
                                                              "Surface terrain","exterieur","pop_active","salaire_moyen","nb_etab_elem"]],df[["Type local encoded"]], test_size=0.30, random_state=42)
```

### Entrainement
#### Arbre de decision 


```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
classifieur_arbre_decision = DecisionTreeClassifier(max_depth=20, min_samples_split=100, min_samples_leaf=40,random_state=42)

# Entraînez le modèle sur l'ensemble d'entraînement
classifieur_arbre_decision.fit(X_train, Y_train)
```

```python
# Faites des prédictions sur l'ensemble de test
predictions = classifieur_arbre_decision.predict(X_test)

# Calculez l'exactitude (accuracy) des prédictions
exactitude = accuracy_score(Y_test, predictions)
```
